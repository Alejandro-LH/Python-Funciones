### 2.1.6 Documentación y Comentarios
#### 2.1.6.1 Uso de docstrings para documentar funciones.


```python
def suma(a, b):
    """
    Suma dos números.

    Parámetros:
    a (int | float): Primer número.
    b (int | float): Segundo número.

    Retorna:
    int | float: Resultado de la suma.
    """
    return a + b

def resta(a, b):
    """
    Resta dos números.

    Parámetros:
    a (int | float): Primer número.
    b (int | float): Segundo número.

    Retorna:
    int | float: Diferencia entre a y b.
    """
    return a - b

def multiplicacion(a, b):
    """Multiplica dos números."""
    return a * b

def division(a, b):
    """Divide dos números."""
    return a / b
```

```python
print(suma.__doc__)
```

#### 2.1.6.2 Comentarios relevantes en el código.


```python
def suma_enteros(a, b):
    # Verifica que ambos argumentos sean enteros
    if not isinstance(a, int) or not isinstance(b, int):
        print("Ambos valores deben ser enteros.")
        return None

    # Calcula la suma
    resultado = a + b

    # Muestra el resultado en pantalla
    print(f"La suma de {a} y {b} es {resultado}.")
    return resultado
```

```python
print(suma_enteros(1, 5))
print(suma_enteros(1.0, 5))
```

### 2.1.8 Pruebas y Depuración
#### 2.1.8.1 Escribir pruebas unitarias para funciones.


```python
import unittest

def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

class TestOperaciones(unittest.TestCase):
    def test_suma(self):
        self.assertEqual(suma(2, 3), 5)
        self.assertEqual(suma(-2, 3), 1)

    def test_resta(self):
        self.assertEqual(resta(2, 3), -1)
        self.assertEqual(resta(5, 2), 3)

unittest.main(argv=[""], exit=False)
```

#### 2.1.8.2 Uso de assertions.


```python
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

assert suma(2, 3) == 5, "debería ser 5"
assert suma(-2, 3) == 1, "debería ser 1"
assert suma(0, 0) == 0, "debería ser 0"
assert resta(2, 3) == -1, "debería ser -1"
assert resta(5, 2) == 3, "debería ser 3"

print("Todas las assertions se cumplieron correctamente.")
```
